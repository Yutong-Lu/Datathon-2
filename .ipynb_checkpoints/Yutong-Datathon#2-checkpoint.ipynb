```python
import pandas as pd
import numpy as np
import math 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve, accuracy_score, precision_score, recall_score
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
from sklearn import neighbors
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Cardiovascular Event Dataset.csv')

# View the first few rows of the dataset
df.head()
```

```python
# Remove identifiers
df=df.drop('id',axis=1)
```

```python
# Check rows with unknown smoking_status
df[df['smoking_status']=="Unknown"] 
# 1544 of them unknown, some are children
```

```python
description = df.describe()
description
```

```python
# only 249/5110 people had a stroke
df[df['stroke']==1]
```

```python
print(df.isnull().sum())
```

```python
# Impute missing values with median
df.fillna(df.median(), inplace=True)
# Checking for missing values
print(df.isnull().sum())
```

```python
# Display pair plot for data visualization
sns.pairplot(df, hue='stroke', palette="magma")
plt.show()
```

```python
# One-hot encode the 'gender' column
gender_dummies = pd.get_dummies(df['gender'])
df = pd.concat([df, gender_dummies], axis=1)
df.head()

# One-hot encode the 'work_type' column
work_dummies = pd.get_dummies(df['work_type'])
df = pd.concat([df, work_dummies], axis=1)
df.head()

# One-hot encode the 'work_type' column
residence_dummies = pd.get_dummies(df['Residence_type'])
df = pd.concat([df, residence_dummies], axis=1)
df.head()

# One-hot encode the 'work_type' column
smoking_dummies = pd.get_dummies(df['smoking_status'])
df = pd.concat([df, smoking_dummies], axis=1)
df.head()
```

```python
# Split the data into features and target variable
X = df.drop(['stroke','gender','ever_married', 'work_type','Residence_type','smoking_status'], axis=1)
y = df['stroke']

# Normalizing data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Training logistic regression model
log_reg = LogisticRegression(penalty='l1', C=1.0, class_weight = 'balanced', solver='liblinear')
log_reg.fit(X_train, y_train)
```

```python
# Use the model to make predictions on the testing data
y_pred = log_reg.predict(X_test)

# Evaluate the model
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print("Confusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(class_report)

# You either have data that does not "contain" enough information to predict your label and your classifier just "guesses" the most frequent class. 
# Or you have so much more data of class 1 than class 0 that the accuracy is better if you allways guess 1 instead of trying to classify correctly.
```

```python
import statsmodels.api as sm
import pandas as pd

# Add a constant term for the intercept
X = sm.add_constant(X)

# Create and fit logistic regression model
log_reg = sm.Logit(y, X).fit()

# Get summary report
summary = log_reg.summary()
print(summary)
```

```python
# apply KNN

max_k = 50

error_rate =[] # list that will store the accuracy score value of k
for i in range (1, max_k):  #Took the range of k from 1 to max_k
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    y_pred =knn.predict(X_test)
    error_rate.append(1 - accuracy_score(y_test, y_pred))

# Plotting the accuracy vs k graph
plt.figure(figsize=(12,6))
plt.plot(range(1, max_k),error_rate,marker="o",
         markerfacecolor="green",
         linestyle="dashed",color="red")
plt.title("Error rate vs k value",fontsize=20)
plt.xlabel("k values",fontsize=20)
plt.ylabel("error rate",fontsize=20)
plt.xticks(range(1, max_k))
plt.show()
```

```python
# choose K based on class I precision

max_k = 50

error_rate =[] # list that will store the accuracy score value of k
for i in range (1, max_k):  #Took the range of k from 1 to max_k
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    y_pred =knn.predict(X_test)
    error_rate.append(precision_score(y_test, y_pred, pos_label=1))

# Plotting the accuracy vs k graph
plt.figure(figsize=(12,6))
plt.plot(range(1, max_k),error_rate,marker="o",
         markerfacecolor="green",
         linestyle="dashed",color="red")
plt.title("Class 1 precision vs k value",fontsize=20)
plt.xlabel("k values",fontsize=20)
plt.ylabel("Class 1 precision",fontsize=20)
plt.xticks(range(1, max_k))
plt.show()
```

```python
# choose K based on class I recall

max_k = 50

error_rate =[] # list that will store the accuracy score value of k
for i in range (1, max_k):  #Took the range of k from 1 to max_k
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    y_pred =knn.predict(X_test)
    error_rate.append(recall_score(y_test, y_pred, pos_label=1))

# Plotting the accuracy vs k graph
plt.figure(figsize=(12,6))
plt.plot(range(1, max_k),error_rate,marker="o",
         markerfacecolor="green",
         linestyle="dashed",color="red")
plt.title("Class 1 recall vs k value",fontsize=20)
plt.xlabel("k values",fontsize=20)
plt.ylabel("Class 1 recall",fontsize=20)
plt.xticks(range(1, max_k))
plt.show()
```

```python
# Apply KNN, the Optimal K was 9 based on the plot

knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(X_train, y_train)

# Predict using the test data
y_pred = knn.predict(X_test)

# Train accuracy
y_train_pred = knn.predict(X_train)
train_accuracy = accuracy_score(y_train, y_train_pred)
print(f"Train Accuracy: {train_accuracy * 100:.2f}%")

# Test accuracy
test_accuracy = accuracy_score(y_test, y_pred)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")

# Evaluate the model
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Split the data into features and target variable
X = df[['age', 'ejection_fraction', 'anaemia', 'high_blood_pressure', 'creatinine_phosphokinase', 'serum_creatinine', 'serum_sodium']]
y = df['DEATH_EVENT']

# Normalizing data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# choose K based on overall accuracy

max_k = 50

error_rate =[] # list that will store the accuracy score value of k
for i in range (1, max_k):  #Took the range of k from 1 to max_k
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    y_pred =knn.predict(X_test)
    error_rate.append(1 - accuracy_score(y_test, y_pred))

# Plotting the accuracy vs k graph
plt.figure(figsize=(12,6))
plt.plot(range(1, max_k),error_rate,marker="o",
         markerfacecolor="green",
         linestyle="dashed",color="red")
plt.title("Error rate vs k value",fontsize=20)
plt.xlabel("k values",fontsize=20)
plt.ylabel("error rate",fontsize=20)
plt.xticks(range(1, max_k))
plt.show()
```

```python
# choose K based on class 1 recall

max_k = round(math.sqrt(299))

recall_rate =[] # list that will store the accuracy score value of k
for i in range (1, max_k):  #Took the range of k from 1 to max_k
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    y_pred =knn.predict(X_test)
    recall_rate.append(recall_score(y_test, y_pred, pos_label=1))

# Plotting the accuracy vs k graph
plt.figure(figsize=(12,6))
plt.plot(range(1, max_k),recall_rate,marker="o",
         markerfacecolor="green",
         linestyle="dashed",color="red")
plt.title("Recall rate vs k value",fontsize=20)
plt.xlabel("k values",fontsize=20)
plt.ylabel("recall rate",fontsize=20)
plt.xticks(range(1, max_k))
plt.show()
```

```python
# choose K based on class 1 precision

max_k = round(math.sqrt(299))

precision_rate =[] # list that will store the accuracy score value of k
for i in range (1, max_k):  #Took the range of k from 1 to max_k
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    y_pred =knn.predict(X_test)
    precision_rate.append(precision_score(y_test, y_pred, pos_label=1))

# Plotting the accuracy vs k graph
plt.figure(figsize=(12,6))
plt.plot(range(1, max_k),precision_rate,marker="o",
         markerfacecolor="green",
         linestyle="dashed",color="red")
plt.title("Precision vs k value",fontsize=20)
plt.xlabel("k values",fontsize=20)
plt.ylabel("precision",fontsize=20)
plt.xticks(range(1, max_k))
plt.show()
```

```python
# Apply KNN, the Optimal K was 3 based on the plot

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# Predict using the test data
y_pred = knn.predict(X_test)

# Train accuracy
y_train_pred = knn.predict(X_train)
train_accuracy = accuracy_score(y_train, y_train_pred)
print(f"Train Accuracy: {train_accuracy * 100:.2f}%")

# Test accuracy
test_accuracy = accuracy_score(y_test, y_pred)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")

# Evaluate the model
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
